```python
#數據可視化
#新增驗證(日期格式、有效日期、時間格式)
#添加錯誤提醒和默認值
import sqlite3 as s
import random as r
import matplotlib.pyplot as plt
from tkinter import ttk       
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
from tkinter import messagebox as msgbox
from PIL import ImageTk,Image
import pandas as pd
import numpy as np
from tkinter.font import Font
from datetime import datetime
from sklearn.linear_model import LinearRegression
import numpy as np
def validate_input(date, hours, minutes, rate):
    """驗證用戶輸入的數據"""
    try:
        # 檢查是否有空值
        if not all([date, hours is not None, minutes is not None, rate]):
            return False, "所有欄位都必須填寫"
        
        if rate == 'select':
            return False, "請選擇評分"
        
        # 驗證日期格式
        date_str = str(date)
        if len(date_str) != 8:
            return False, "日期必須為8位數字 (YYYYMMDD)"
        
        year = int(date_str[:4])
        month = int(date_str[4:6])
        day = int(date_str[6:])
        
        # 檢查日期是否有效
        try:
            datetime(year, month, day)
        except ValueError:
            return False, "無效的日期"
        
        # 驗證時間
        if not (0 <= hours <= 24):
            return False, "小時必須在0-24之間"
        if not (0 <= minutes <= 59):
            return False, "分鐘必須在0-59之間"
        
        return True, ""
    except:
        return False, "輸入格式錯誤"

def n_r(): #new record
    global n1,n2_h,n2_m,n3,combobox1,w_1
    w_1=tk.Toplevel()
    w_1.title('NEW record')
    w_1.geometry('570x300')
    n1=tk.IntVar() #DATE
    n2_h=tk.IntVar() #DURATION HOUR
    n2_m=tk.IntVar() #DURATION HOUR
    n3=tk.StringVar() #NOTE
    
    # 設置默認值
    n2_h.set(0)
    n2_m.set(0)
    
    tk.Label(w_1, text='Date',font='14').place(x=10,y=10)
    tk.Entry(w_1, width=14, textvariable=n1).place(x=60,y=12)
    tk.Label(w_1, text='Duration',font='14').place(x=180,y=10)
    tk.Entry(w_1, width=5, textvariable=n2_h).place(x=252,y=12)
    tk.Label(w_1, text='hr',font='14').place(x=292,y=10)
    tk.Entry(w_1, width=7, textvariable=n2_m).place(x=312,y=12)
    tk.Label(w_1, text='min',font='14').place(x=352,y=10)
    tk.Label(w_1, text='Rate',font='14').place(x=10,y=40)
    values=['Awsome!','Great','Not Bad','Bad','Terrible!']
    combobox1=ttk.Combobox(w_1, values=values,width=8)
    combobox1.place(x=75,y=42)
    combobox1.set('select')
    tk.Label(w_1, text='Note',font='14').place(x=10,y=70)
    tk.Entry(w_1, width=60, textvariable=n3).place(x=10,y=100)
    n3.set('enter notes')
    tk.Button(w_1, text='Done',font='20',command=new).place(x=250,y=200)

def new():
    try:
        new_1 = n1.get()
        new_h = n2_h.get()
        new_m = n2_m.get()
        newcom_1 = combobox1.get()
        
        # 驗證輸入
        valid, message = validate_input(new_1, new_h, new_m, newcom_1)
        if not valid:
            msgbox.showerror('錯誤', message)
            return
        
        new_t = (60 * new_h) + new_m
        new_3 = n3.get()
        
        ans = msgbox.askokcancel('Add a new record?','click to confirm',icon='info')
        if ans:
            conn = s.connect("D:\DB\sleeping recorder.db")
            x = (new_1, new_t, newcom_1, new_3)
            sql = 'insert into data values(?,?,?,?)'
            conn.execute(sql,x)
            conn.commit()
            conn.close() 
            
            # 清空輸入
            n1.set(0)
            n2_h.set(0)
            n2_m.set(0)
            n3.set('enter notes')
            combobox1.set('select')
            
            msgbox.showinfo('成功', '記錄已添加')
            w_1.destroy()
            
    except Exception as e:
        msgbox.showerror('錯誤', f'發生錯誤：{str(e)}')

def m_r(): #modify record
    global m1,m2_h,m2_m,m3,mcombobox1,w_m
    w_m=tk.Toplevel()
    w_m.title('MODIFY record')
    w_m.geometry('600x300')
    m1=tk.IntVar() #DATE
    m2_h=tk.IntVar() #DURATION HOUR
    m2_m=tk.IntVar() #DURATION HOUR
    m3=tk.StringVar() #NOTE
    
    # 設置默認值
    m2_h.set(0)
    m2_m.set(0)
    
    tk.Label(w_m, text='Date',font='14').place(x=10,y=10)
    tk.Entry(w_m, width=14, textvariable=m1).place(x=60,y=12)
    tk.Label(w_m, text='Duration',font='14').place(x=180,y=10)
    tk.Entry(w_m, width=5, textvariable=m2_h).place(x=252,y=12)
    tk.Label(w_m, text='hr',font='14').place(x=292,y=10)
    tk.Entry(w_m, width=7, textvariable=m2_m).place(x=312,y=12)
    tk.Label(w_m, text='min',font='14').place(x=352,y=10)
    tk.Label(w_m, text='Rate',font='14').place(x=10,y=40)
    values=['Awsome!','Great','Not Bad','Bad','Terrible!']
    mcombobox1=ttk.Combobox(w_m, values=values,width=8)
    mcombobox1.place(x=75,y=42)
    mcombobox1.set('select')
    tk.Label(w_m, text='Note',font='14').place(x=10,y=70)
    tk.Entry(w_m, width=60, textvariable=m3).place(x=10,y=100)
    m3.set('enter notes')
    tk.Button(w_m, text='MODIFY',font='20',command=modify).place(x=250,y=200)

def modify():
    try:
        mod_1 = m1.get()
        mod_h = m2_h.get()
        mod_m = m2_m.get()
        modcom_1 = mcombobox1.get()
        
        # 驗證輸入
        valid, message = validate_input(mod_1, mod_h, mod_m, modcom_1)
        if not valid:
            msgbox.showerror('錯誤', message)
            return
            
        mod_t = (60 * mod_h) + mod_m
        mod_3 = m3.get()
        
        ans = msgbox.askokcancel('MODIFY the record','confirm to modify date',icon='info')
        if ans:
            conn = s.connect("D:\DB\sleeping recorder.db")
            x = (mod_t, modcom_1, mod_3, mod_1)
            sql = 'update data set time=?,rate=?,Notes=? where date=?'
            cursor = conn.execute(sql,x)
            
            if cursor.rowcount == 0:
                msgbox.showwarning('警告', '未找到要修改的記錄')
            else:
                conn.commit()
                msgbox.showinfo('成功', '記錄已修改')
                m3.set('enter notes')
                mcombobox1.set('select')
                w_m.destroy()
            
            conn.close()
            
    except Exception as e:
        msgbox.showerror('錯誤', f'發生錯誤：{str(e)}')

def d_r(): #delete record
    global d_1,w_3
    w_3=tk.Toplevel()
    w_3.title('delete record')
    w_3.geometry('600x300')
    d_1=tk.IntVar()
    tk.Label(w_3, text='Date',font='14').place(x=10,y=10)
    tk.Entry(w_3, width=14, textvariable=d_1).place(x=80,y=12)
    tk.Button(w_3,text ='DELETE',width='14',command=del_w).place(x=180,y=80)
    w_3.mainloop()
    
def del_w():
    global d_1,w_3
    ans=msgbox.askokcancel('DELETE the record','confirm to delete date permanently',icon='info')
    if ans==True:
        del_1=d_1.get()
        conn=s.connect("D:\DB\sleeping recorder.db")
        conn.execute('delete from data where date=?',(del_1,))
        conn.commit()
        conn.close()
        d_1.set(0)
        w_3.destroy()

def q_d():
    global w_4,q_d1,q_d2,tree,q_da
    w_4=tk.Toplevel()
    w_4.title('Search by DATE')
    w_4.geometry('600x600')
    q_d1=tk.IntVar()
    q_d2=tk.IntVar()
    tk.Label(w_4, text='開始日期',font='12').place(x=30,y=30)
    tk.Entry(w_4, width=10, textvariable=q_d1).place(x=120,y=32)
    tk.Label(w_4, text='結束日期',font='12').place(x=230,y=30)
    tk.Entry(w_4, width=10, textvariable=q_d2).place(x=320,y=32)
    tk.Button(w_4,text ='查詢',width='15', command=q_da).place(x=420,y=30)
    tree=ttk.Treeview(w_4,show='headings',column=('Date','Time','Rate','Notes'))
    tree.column('Date', anchor='center', width=100)
    tree.column('Time', anchor='center', width=90)
    tree.column('Rate', anchor='center', width=70)
    tree.column('Notes', anchor='center', width=250)
    tree.heading('Date', text='Date')
    tree.heading('Time', text='Time')
    tree.heading('Rate', text='Rate')
    tree.heading('Notes', text='Notes')
    tree.place(x=40,y=100,height=450)
    w_4.mainloop()

def q_da():
    global w_4,q_d1,q_d2,tree,q_da
    q_1=q_d1.get()
    q_2=q_d2.get()
    conn=s.connect("D:\DB\sleeping recorder.db")
    results=conn.execute('select * from data where date>= ? and date<=?',(q_1,q_2)) 
    i=0
    for record in results:
        tree.insert(parent='', index=i, iid=i, text='', values=(record[0],record[1],record[2],record[3]))
        i=i+1
    conn.close()

def plot_sleep_data():
    w_plot = tk.Toplevel()
    w_plot.title('睡眠時間統計圖')
    w_plot.geometry('800x600')

    try:
        # 連接數據庫
        conn = s.connect("D:/DB/sleeping recorder.db")
        cursor = conn.cursor()
        
        # 直接執行 SQL 查詢並印出結果
        cursor.execute("SELECT * FROM data")
        records = cursor.fetchall()
        print("原始數據：", records)  # 檢查是否有數據
        
        if not records:
            msgbox.showinfo('提示', '數據庫中沒有記錄')
            return
        
        # 使用 pandas 讀取數據
        df = pd.read_sql_query("""
            SELECT date, time 
            FROM data 
            ORDER BY date
        """, conn)
        print("DataFrame 內容：\n", df)  # 檢查 DataFrame
        
        if df.empty:
            msgbox.showinfo('提示', '無法讀取數據')
            return
        
        # 數據處理
        df['hours'] = df['time'].astype(float) / 60  # 確保轉換為浮點數
        df['date_str'] = df['date'].astype(str)
        
        print("處理後的數據：")
        print("日期列表：", df['date_str'].tolist())
        print("小時列表：", df['hours'].tolist())
        
        # 創建圖表
        fig = Figure(figsize=(10, 6))
        ax = fig.add_subplot(111)
        
        # 繪製數據點和線
        ax.plot(range(len(df)), df['hours'], 'bo-', label='睡眠時間')  # 藍色點和線
        
        # 設置標籤
        ax.set_title('睡眠時間趨勢圖', fontsize=14, pad=20)
        ax.set_xlabel('日期', fontsize=12)
        ax.set_ylabel('睡眠時數', fontsize=12)
        
        # 設置 x 軸刻度和標籤
        ax.set_xticks(range(len(df)))
        ax.set_xticklabels(df['date_str'], rotation=45)
        
        # 設置 y 軸範圍
        y_max = df['hours'].max()
        if not pd.isna(y_max):  # 確保有有效的最大值
            ax.set_ylim(0, y_max + 1)
        
        # 添加網格
        ax.grid(True, linestyle='--', alpha=0.7)
        
        # 添加圖例
        ax.legend()
        
        # 調整布局
        fig.tight_layout()
        
        # 顯示圖表
        canvas = FigureCanvasTkAgg(fig, master=w_plot)
        canvas.draw()
        canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True, padx=10, pady=10)
        
        conn.close()

    except Exception as e:
        msgbox.showerror('錯誤', f'生成圖表時發生錯誤：{str(e)}')
        print(f"錯誤詳情：{str(e)}")
        import traceback
        traceback.print_exc()

from sklearn.linear_model import LinearRegression
import numpy as np

# 在其他函數（如 plot_sleep_data）之後，main 之前添加：
def predict_sleep():
    w_predict = tk.Toplevel()
    w_predict.title('睡眠時間預測')
    w_predict.geometry('400x500')
    
    global predict_date
    predict_date = tk.IntVar()
    
    tk.Label(w_predict, text='請輸入要預測的日期 (YYYYMMDD):', font='14').place(x=20, y=20)
    tk.Entry(w_predict, width=14, textvariable=predict_date).place(x=20, y=50)
    
    result_label = tk.Label(w_predict, text='', justify=tk.LEFT, font=('Courier', 12))
    result_label.place(x=20, y=120)
    
    canvas_frame = tk.Frame(w_predict)
    canvas_frame.place(x=20, y=250, width=360, height=200)
    
    def do_predict():
        try:
            conn = s.connect("D:/DB/sleeping recorder.db")  # 使用您的數據庫路徑
            query = "SELECT date, time FROM data ORDER BY date"
            df = pd.read_sql_query(query, conn)
            conn.close()
            
            if len(df) < 2:
                msgbox.showwarning('警告', '數據量不足，無法進行預測')
                return
                
            X = df['date'].values.reshape(-1, 1)
            y = df['time'].values / 60
            plt.rcParams['font.family']='Microsoft JhengHei'
            model = LinearRegression()
            model.fit(X, y)
            
            input_date = predict_date.get()
            predicted_hours = model.predict([[input_date]])[0]
            
            mean_hours = y.mean()
            std_hours = y.std()
            lower_bound = max(0, predicted_hours - std_hours)
            upper_bound = predicted_hours + std_hours
            
            result_text = f'''
預測日期: {input_date}
預測睡眠時間: {predicted_hours:.2f} 小時
預測區間: {lower_bound:.2f} - {upper_bound:.2f} 小時
平均睡眠時間: {mean_hours:.2f} 小時
'''
            result_label.config(text=result_text)
            
            fig = Figure(figsize=(8, 4))
            ax = fig.add_subplot(111)
            ax.scatter(X, y, color='blue', label='歷史數據')
            
            X_range = np.array([min(X)[0], input_date, max(X)[0]]).reshape(-1, 1)
            y_pred = model.predict(X_range)
            ax.plot(X_range, y_pred, color='red', label='趨勢線')
            plt.rcParams['font.family']='Microsoft JhengHei'
            ax.scatter([input_date], [predicted_hours], color='green', s=100, label='預測值')
            
            ax.set_title('睡眠時間預測')
            ax.set_xlabel('日期')
            ax.set_ylabel('睡眠時數')
            ax.legend()
            
            for widget in canvas_frame.winfo_children():
                widget.destroy()
            
            canvas = FigureCanvasTkAgg(fig, master=canvas_frame)
            canvas.draw()
            canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
            
        except Exception as e:
            msgbox.showerror('錯誤', f'預測時發生錯誤：{str(e)}')
            print(f"錯誤詳情：{str(e)}")
    
    tk.Button(w_predict, text='預測', command=do_predict, font='14').place(x=20, y=80)
# 在主窗口的菜單設置部分，修改為：

win = tk.Tk()
win.title('sleeping recorder')
win.geometry('600x600')
menu = tk.Menu(win)
win.config(menu=menu)

# 維護菜單
maintain_menu = tk.Menu(menu, tearoff=False)
menu.add_cascade(label="update", menu=maintain_menu)
maintain_menu.add_command(label="NEW record", command=n_r)
maintain_menu.add_command(label="MODIFY record", command=m_r)
maintain_menu.add_command(label="DELETE record", command=d_r)

# 查詢菜單
query_menu = tk.Menu(menu, tearoff=False)
menu.add_cascade(label="query", menu=query_menu)
query_menu.add_command(label="Search by DATE", command=q_d)
query_menu.add_command(label="睡眠時間統計圖", command=plot_sleep_data)

# 預測菜單
predict_menu = tk.Menu(menu, tearoff=False)  # 先創建預測菜單
menu.add_cascade(label="預測", menu=predict_menu)  # 然後添加到主菜單
predict_menu.add_command(label="睡眠時間預測", command=predict_sleep)  # 最後添加命令

# 結束菜單
menu.add_cascade(label="END", command=win.destroy)

win.mainloop()
```
